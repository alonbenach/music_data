```python
import requests
import spotipy

# Prepare to make a request to Spotify server
url = 'https://accounts.spotify.com/api/token'
client_id= '7c2c77f7146a408bbd8aade210c853c1'
client_secret = '0e92d709e3b4442fac06b8c58b17bbfd'

auth_response = requests.post(url, {
    'grant_type':'client_credentials',
    'client_id':client_id,
    'client_secret':client_secret,
})

# convert the response to JSON
auth_response_data = auth_response.json()

# save the access token
access_token = auth_response_data['access_token']

access_token
```

```python
# Format a get request
headers = {
    'Authorization': 'Bearer {token}'.format(token=access_token)
}
# base URL of all Spotify API endpoints
BASE_URL = 'https://api.spotify.com/v1/'

# Track ID from the URI
track_id = '6y0igZArWVi6Iz0rj35c1Y'
```

```python
sp = spotipy.Spotify(auth=headers)

artist= input("Please enter artist name: ")
track= input("Please enter track name: ")

track_id = sp.search(q='artist:' + artist + ' track:' + track, type='track')
track_id
```

```python
trackId = track_id['tracks']['items'][0]['id']
trackId
```

```python
# actual GET request with proper header
r = requests.get(BASE_URL + 'audio-features/' + trackId, headers=headers)
r = r.json()
r
```
